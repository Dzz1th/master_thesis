```python
import requests
from bs4 import BeautifulSoup

def extract_minutes_text(url):
    # Send an HTTP GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Check for request errors

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Create a set to store unique paragraphs
    seen_paragraphs = set()
    
    # Find all <p> tags and extract their text
    paragraphs = []
    for p in soup.find_all('p'):
        text = p.get_text(strip=True)
        # Check if the paragraph has already been seen
        if text and text not in seen_paragraphs:
            seen_paragraphs.add(text)
            paragraphs.append(text)
    
    return '\n'.join(paragraphs)  # Return the extracted paragraphs

import re
from datetime import datetime

def extract_date_from_url(url):
    # Regular expression to capture the 8-digit date (YYYYMMDD) from the URL
    date_match = re.search(r'(\d{8})\.htm', url)
    if date_match:
        date_str = date_match.group(1)
        try:
            date = datetime.strptime(date_str, "%Y%m%d")
            return date.strftime("%Y-%m-%d")
        except ValueError:
            return "Invalid date format"
    
    return "Date not found in the URL"
```

```python
from tqdm import tqdm

minutes_texts = {'Date': [], 'Text': []}

for url in tqdm(urls):
    try:
        date = extract_date_from_url(url)
        minutes_text = extract_minutes_text(url)
        minutes_texts['Date'].append(date)
        minutes_texts['Text'].append(minutes_text)
    except BaseException as exc:
        print('Url : ', url, 'Exception: ', exc)

```

```python
import pandas as pd
import os 



df = pd.read_csv("Fed_Scrape-2015-2023.csv")

df = df[df['Type'] == 1][['Date', 'Text']]
df = df.groupby('Date').agg({
    'Text': lambda x: ' '.join(x)
}).reset_index()

df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')

import os

# Define the directory path
directory_path = '/Users/dzz1th/Job/mgi/Soroka/data'

# Create an empty dictionary to store the data
data_dict = {'Date': [], 'Text': []}

# Iterate over the files in the directory
for filename in os.listdir(directory_path):
    # Check if the file is a text file and has a date in the format YYYYMMDD
    if filename.endswith('.txt') and len(filename) == 12:
        # Extract the date from the file name
        date = filename[:-4]  # remove the .txt extension
        # Read the text from the file
        with open(os.path.join(directory_path, filename), 'r') as file:
            text = file.read()
        data_dict['Date'].append(date)
        data_dict['Text'].append(text)

add_df = pd.DataFrame.from_dict(data_dict, orient='columns')
add_df['Date'] = pd.to_datetime(add_df['Date'], format='%Y%m%d')

df = pd.concat([add_df, df])
df = df.reset_index(drop=True)
df = df.drop_duplicates(subset=['Date'], keep='first')
df = df.sort_values(by='Date')

df.to_csv('FED_Scrape-2000-2024.csv')



```
