```python
series_ids = {
        "Capacity Utilization: Manufacturing": "TCU",
        "Real Personal Income": "W875RX1",
        "GDP": "GDPC1",
        "Unemployment": "UNRATE",
        "Labor Participation": "CIVPART",
        "Average Hourly Earnings": "AHETPI",
        "CPI": "CPIAUCSL",
        "PCE": "PCEPILFE",
        "Durable Goods Orders": "DGORDER",
        "Oil Price": "DCOILWTICO",
        "S&P 500": "SP500",
        "Housing Starts": "HOUST",
        "Building Permits": "PERMIT",
        "Consumer Confidence": "UMCSENT",
        "Job Openings": "JTSJOL",
        "Wage Growth (ECI: Wages & Salaries)": "ECIWAG",
        "Nonfarm Payrolls": "PAYEMS"
    }

def preprocess_data_for_sentiment_prediction(macro_df, sentiment_dates, macro_column: str, pct_change: bool = True):
    macro_df = macro_df.sort_index()
    conf_dates = sorted(sentiment_dates)

    macro_changes = pd.DataFrame()
    for i in range(1, len(conf_dates)):
        current_date = conf_dates[i]
        previous_date = conf_dates[i-1]
        
        # Create a row for this pair of dates
        row_data = {}
        row_data['conf_date'] = current_date
        row_data['previous_conf_date'] = previous_date
        
        # Get the latest available value for each macro variable before each conference date
        for column in macro_df.columns:
            # Filter data between the two dates
            period_data = macro_df.loc[previous_date:current_date, column]
            
            # Remove NaN values (keeping only release dates with actual values)
            period_data = period_data.dropna()
            
            # If we have data points in this period
            if not period_data.empty:
                # Most recent value before current conference
                if current_date in period_data.index:
                    current_val = period_data.loc[current_date]
                else:
                    current_val = period_data[period_data.index < current_date]
                    current_val = current_val.iloc[-1] if not current_val.empty else np.nan
                
                # Most recent value before previous conference
                if previous_date in period_data.index:
                    previous_val = period_data.loc[previous_date]
                else:
                    previous_val = period_data[period_data.index < previous_date]
                    previous_val = previous_val.iloc[-1] if not previous_val.empty else np.nan
                
                # Calculate absolute change
                if not (np.isnan(current_val) or np.isnan(previous_val)):
                    row_data[f'{column}_change'] = current_val - previous_val
                    row_data[f'{column}_pct_change'] = (current_val - previous_val) / previous_val * 100
                    row_data[f'{column}_current'] = current_val
                    row_data[f'{column}_previous'] = previous_val
            
        # Add this row to our results dataframe
        macro_changes = pd.concat([macro_changes, pd.DataFrame([row_data])], ignore_index=True)
    
    # 5. Merge with sentiment data
    merged_df = pd.merge(
        macro_changes, 
        sentiment_df, 
        left_on='conf_date', 
        right_index=True,
        how='inner'
    )
    
    return merged_df

# Example usage:
# processed_data = preprocess_data_for_sentiment_prediction(macro_df, sentiment_df)
```
